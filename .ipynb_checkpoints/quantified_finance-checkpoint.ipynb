```python
# 引入数据处理的核心库
import pandas as pd  # Excel的超级加强版，量化核心
import numpy as np   # 矩阵运算，C++底层，速度极快
import matplotlib.pyplot as plt # 画图工具

# 让图片直接显示在网页上
%matplotlib inline 

print("环境配置成功，准备起飞！")
```

```python
# 1. 生成时间索引 (模拟从2024年1月1日开始的250个交易日)
dates = pd.date_range(start='2024-01-01', periods=250, freq='B') # 'B'代表Business Day(工作日)

# 2. 模拟股价走势 (随机游走)
# np.random.randn(250): 生成250个服从标准正态分布的随机数
# .cumsum(): 累加求和，模拟价格波动
# + 100: 假设初始股价是 100 块
price_data = np.random.randn(250).cumsum() + 100 

# 3. 装载进 DataFrame (相当于Excel的一张表)
df = pd.DataFrame(data=price_data, index=dates, columns=['Close'])

# 查看前5行数据
df.head()
```
